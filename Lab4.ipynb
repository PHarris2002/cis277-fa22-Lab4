## NFL Offense Data

1. Import the NFL week 3, 4, and 5 data and add a week column to dataset and set the value to 3, 4, or 5 for all rows from that week.

```python
nfl_offense_week3_df = pd.read_csv("nfloffenseweek3.csv")
nfl_offense_week4_df = pd.read_csv("nfloffenseweek4.csv")
nfl_offense_week5_df = pd.read_csv("nfloffenseweek5.csv")
```

Adding week column to week 3

```python
NFL_OW3_df2 = nfl_offense_week3_df.assign(week=3)
NFL_OW3_df2.loc[:, ["team", "week"]]
```

Adding week column to week 4

```python
NFL_OW4_df2 = nfl_offense_week4_df.assign(week=4)
NFL_OW4_df2.loc[:, ["team", "week"]]
```

Adding week column to week 5

```python
NFL_OW5_df2 = nfl_offense_week3_df.assign(week=5)
NFL_OW5_df2.loc[:, ["team", "week"]]
```

2. Combine the weekly dataframes by adding the dataframes together row-wise.

```python
combined_data = pd.concat([NFL_OW3_df2, NFL_OW4_df2, NFL_OW5_df2], ignore_index=True)
combined_data
```

3. Make up 1 interesting question about this dataset. Answer it.

Question: What is the difference number in total yards played for all offense teams between Week 3 and Week 4?

```python
week_3_mask = combined_data["week"] == 3
week_4_mask = combined_data["week"] == 4

week_3_total_yard_sum = combined_data[week_3_mask].total_yards.sum()
week_4_total_yard_sum = combined_data[week_4_mask].total_yards.sum()

if week_3_total_yard_sum > week_4_total_yard_sum:
    difference = week_3_total_yard_sum - week_4_total_yard_sum
else:
    difference = week_4_total_yard_sum - week_3_total_yard_sum

print(f"The amount of total yards for all teams between weeks 3 and 4 is {difference} yards")
```
